```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the .ener file (update the path as needed)
file_path = 'Outputs/NPT-1.ener'

# Read the .ener file, assuming space or tab-separated columns, and that the file has a header.
# If your file does not have a header, you can add header=None and provide column names manually.

data = pd.read_csv(file_path, delim_whitespace=True, header=0)


# Extract column 1 (time) and column 4 (potential energy)
time = data.iloc[:, 1]  # First column (Time)
potential_energy = data.iloc[:, 4]  # Fourth column (Potential Energy)

# Plot the potential energy over time
plt.figure(figsize=(10, 6))
plt.plot(time, potential_energy, label='NPT Potential Energy', color='r')

plt.xlabel('Time (fs)')
plt.ylabel('Potential Energy (a.u.)')
plt.title('NPT Potential Energy vs Time')
plt.legend(loc='best')
plt.grid(True)
# Save the plot as a .jpg file
plt.savefig('NPT_potential_energy.jpg', format='jpg')

plt.show()
```
