```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import interp1d

# Load the first dataset (MONITOR-COLVAR.metadynLog) that goes to 30,000 fs
file_path_1 = 'Outputs/metad1_initmonitor/METADYN-COLVAR.metadynLog'
data_1 = pd.read_csv(file_path_1, delim_whitespace=True, header=None)

# Extract time and reaction coordinate from the first file
time_1 = data_1[0]  # Time column from first file
reaction_coordinate_1 = data_1[1]  # Collective variable column from first file

# Load the second dataset that goes to 5,000 fs
file_path_2 = 'Outputs/monitor/MONITOR-COLVAR.metadynLog'  # Replace with the actual path
data_2 = pd.read_csv(file_path_2, delim_whitespace=True, header=None)

# Extract time and reaction coordinate from the second file
time_2 = data_2[0]  # Time column from second file
reaction_coordinate_2 = data_2[1]  # Collective variable column from second file

# Load the third dataset that goes to 30 ps (30,000 fs)
file_path_3 = 'Outputs/metad1_initmonitor_freqhills/METADYN-COLVAR.metadynLog'  # Replace with the actual path
data_3 = pd.read_csv(file_path_3, delim_whitespace=True, header=None)

# Extract time and reaction coordinate from the third file
time_3 = data_3[0]  # Time column from third file
reaction_coordinate_3 = data_3[1]  # Collective variable column from third file

# Rescale the time of the second dataset to match the first one (30,000 fs)
time_2_rescaled = np.linspace(0, time_1.max(), len(time_2))

# Rescale the time of the third dataset to match the first one (30,000 fs)
time_3_rescaled = np.linspace(0, time_1.max(), len(time_3))


# Plot all three datasets on the same graph
plt.figure(figsize=(10, 6))

# Plot the first dataset (Reaction Coordinate 1)
plt.plot(time_1, reaction_coordinate_1, label='Metadynamics 30 ps, Hill Height: 1.5 kcal/mol, Hill Width: 0.5, Hill Frequency: 100', color='blue')

# Plot the second dataset (Reaction Coordinate 2) after rescaling the time
plt.plot(time_2_rescaled, reaction_coordinate_2, label='Standard QMMM 5ps, Hill Height: 1.5 kcal/mol, Hill Width: 0.5, Hill Frequency: 100', color='red')

# Plot the third dataset (Reaction Coordinate 3) after rescaling the time
plt.plot(time_3_rescaled, reaction_coordinate_3, label='Metadynamics_freqhills 30 ps, Hill Height: 1.5 kcal/mol, Hill Width: 0.5, Hill Frequency: 20', color='green')

# Labeling the axes and the title
plt.xlabel('Time (fs)')  # Adjust units if necessary
plt.ylabel('Difference Between C-O Bond Breaking and C-C Bond Forming')
plt.title('Comparison of Reaction Coordinates vs Time')

# Adding grid and legend
plt.grid(True)
plt.legend()

# Show the plot
plt.savefig("Reaction_Coordinate_Comparison.jpg", format="jpg")
plt.show()
```
