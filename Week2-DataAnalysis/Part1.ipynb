```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
 data = pd.read_csv("1hz3_T310.run.25000000.energy.xvg", skiprows=29, delim_whitespace=True, header=None)
```

```python
data
```

```python
# Plot column 1 vs column 0
plt.figure(figsize=(14, 6))
plt.plot(data[0], data[1], label='Potential Energy vs Time')
plt.xlabel('Time (ps)')  # Label for the X-axis
plt.ylabel('Potential Energy (kJ/mol)')  # Label for the Y-axis
plt.title('Potential Energy vs Time')  # Title of the plot

plt.grid(True)
plt.show()
```

```python
# Plot column 1 vs column 0
plt.figure(figsize=(14, 6))
plt.plot(data[0], data[2], label='Kinetic Energy vs Time', color='tab:red')
plt.xlabel('Time (ps)')  # Label for the X-axis
plt.ylabel('Potential Energy (kJ/mol)')  # Label for the Y-axis
plt.title('Kinetic Energy vs Time')  # Title of the plot

plt.grid(True)
plt.show()
```

```python
# Plot column 1 vs column 0
plt.figure(figsize=(14, 6))
plt.plot(data[0], data[3], label='Total Energy vs Time', color='tab:purple')
plt.xlabel('Time (ps)')  # Label for the X-axis
plt.ylabel('Potential Energy (kJ/mol)')  # Label for the Y-axis
plt.title('Total Energy vs Time')  # Title of the plot

plt.grid(True)
plt.show()
```

```python
# Plot column 1 vs column 0
plt.figure(figsize=(14, 6))
plt.plot(data[0], data[4], label='Temperature vs Time', color='tab:orange')
plt.xlabel('Time (ps)')  # Label for the X-axis
plt.ylabel('Temperature (K)')  # Label for the Y-axis
plt.title('Temperature vs Time')  # Title of the plot

plt.grid(True)
plt.show()
```

```python
# Plot column 1 vs column 0
plt.figure(figsize=(14, 6))
plt.plot(data[0], data[5], label='Pressure vs Time', color='tab:pink')
plt.xlabel('Time (ps)')  # Label for the X-axis
plt.ylabel('Pressure (bar)')  # Label for the Y-axis
plt.title('Pressure vs Time')  # Title of the plot

plt.grid(True)
plt.show()
```

```python
# Plot column 1 vs column 0
plt.figure(figsize=(14, 6))
plt.plot(data[0], data[6], label='Volume vs Time', color='tab:green')
plt.xlabel('Time (ps)')  # Label for the X-axis
plt.ylabel('Volume (nm^3)')  # Label for the Y-axis
plt.title('Volume vs Time')  # Title of the plot


plt.grid(True)
plt.show()
```

```python
# Create a figure and axis
fig, ax1 = plt.subplots(figsize=(14, 8))

# Plot column 1 (Y1) with the left y-axis
ax1.plot(data[0], data[4], label='Column 1', color='blue')
ax1.set_xlabel('Time (ps)')
ax1.set_ylabel('Temperature (K)', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# Create a second y-axis sharing the same x-axis
ax2 = ax1.twinx()

# Plot column 2 (Y2) with the right y-axis
ax2.plot(data[0], data[2], label='Column 2', color='purple')
ax2.set_ylabel('Kinetic Energy (kJ/mol)', color='purple')
ax2.tick_params(axis='y', labelcolor='purple')

# Set title and grid
plt.title('Temperature and Kinetic Energy vs. Time')
ax1.grid(True)

# Show the plot
plt.show()
```
