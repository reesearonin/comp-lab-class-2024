```python
import mdtraj as md
import numpy as np
import matplotlib.pyplot as plt

# Load your trajectory and topology
traj = md.load('concatenated.xtc', top='step5_5.gro')

# Define the indices for the first and last atoms
first_atom_index = 0  # Adjust as necessary
last_atom_index = traj.n_atoms - 1  # Last atom

# Prepare arrays to store data
time = traj.time
end_to_end_distance = np.zeros(traj.n_frames)
radius_of_gyration = np.zeros(traj.n_frames)

# Calculate end-to-end distance and radius of gyration
for i in range(traj.n_frames):
    end_to_end_distance[i] = md.compute_distances(traj[i], [[first_atom_index, last_atom_index]])[0][0]
    radius_of_gyration[i] = md.compute_rg(traj[i])

# Create plots
plt.figure(figsize=(12, 5))

# Plot End-to-End Distance
plt.subplot(1, 2, 1)
plt.plot(time, end_to_end_distance, label='End-to-End Distance', color='blue')
plt.title('End-to-End Distance vs Time of Art-Est with Water')
plt.xlabel('Time (ps)')
plt.ylabel('End-to-End Distance (nm)')
plt.legend()
plt.grid()

# Save End-to-End Distance plot as PNG
plt.savefig('end_to_end_distance_With_Water.png', dpi=300)

# Plot Radius of Gyration
plt.subplot(1, 2, 2)
plt.plot(time, radius_of_gyration, label='Radius of Gyration', color='orange')
plt.title('Radius of Gyration vs Time of Art-Est with Water')
plt.xlabel('Time (ps)')
plt.ylabel('Radius of Gyration (nm)')
plt.legend()
plt.grid()

# Save Radius of Gyration plot as PNG
plt.savefig('radius_of_gyration_With_Water.png', dpi=300)

# Show plots
plt.tight_layout()
plt.show()
```
