```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Define a function to read the .ener file and extract the time and energy
def read_ener_file(filename):
    """
    Reads the .ener file and extracts time and energy columns.
    
    Parameters:
    filename (str): Path to the .ener file.
    
    Returns:
    time (ndarray): Array of time values.
    energy (ndarray): Array of energy values.
    """
    # Read the data from the .ener file
    data = np.loadtxt(filename, comments='#')  # Assumes that the file has a header or comments
    
    # Extract time and energy columns
    # Adjust column indices based on the structure of the .ener file
    time = data[:, 1]  # Assuming time is in the first column
    energy = data[:, 2]  # Assuming energy is in the second column
    
    return time, energy

# Load your .ener file (example: 'simulation_data.ener')
filename = 'Water/water_aimd_1ns_every0.5fs-1.ener'
time, energy = read_ener_file(filename)

# Plot Energy vs Time
plt.figure(figsize=(10, 6))
plt.plot(time, energy, label="Energy vs Time", color='b')
plt.title("Kinetic Energy vs Time for Water")
plt.xlabel("Time (fs)")  # Adjust units based on your data
plt.ylabel("Energy (a.u.)")  # Adjust units based on your data
plt.grid(True)
plt.legend()
plt.savefig("Water_Energy.jpg", format="jpg")
plt.show()
```

```python
import MDAnalysis as mda
import numpy as np
import matplotlib.pyplot as plt
from MDAnalysis.analysis import rdf

# Load the XYZ file (replace with the path to your actual file)
xyz_file = 'Runs/Water/water_aimd_1ns_every0.5fs-pos-1.xyz'  # Replace with the path to your .xyz file

# Create the Universe from the XYZ file (assume it's a single-frame trajectory)
u = mda.Universe(xyz_file)

# Set the box size manually (since the box is cubic, the same value for all three dimensions)
box_size = np.array([12.42, 12.42, 12.42])  # 12.42 Å unit cell dimension for each axis

# Set the unit cell dimensions in the Universe object
u.dimensions = np.array([12.42, 12.42, 12.42, 90.0, 90.0, 90.0])  # Cubic box with 90° angles

# Select oxygen atoms (assuming atom name is 'O')
oxygens = u.select_atoms('name O')

# Compute the RDF for the oxygen atoms
rdf_analysis = rdf.InterRDF(oxygens, oxygens, range=(0.0, 2.5), bins=100)
rdf_analysis.run()

# Plot the radial distribution function
plt.plot(rdf_analysis.bins, rdf_analysis.rdf, label='g(r) for Oxygens')
plt.xlabel('r (nm)')
plt.ylabel('g(r)')
plt.title('Radial Distribution Function (g(r)) for Oxygens for Water')
plt.grid(True)
plt.legend()
plt.savefig("Water_Radial_Distribution.jpg", format="jpg")
plt.show()
```
